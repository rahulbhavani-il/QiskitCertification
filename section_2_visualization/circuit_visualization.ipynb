## `plot_circuit_layout()` - Transpilation Qubit Mapping

### Signature
```python
plot_circuit_layout(circuit, backend, view='virtual', qubit_coordinates=None)
```

### Parameters
| Parameter | Type | Required | Default | Description |
|-----------|------|----------|---------|-------------|
| `circuit` | `QuantumCircuit` | Yes | - | **MUST be transpiled circuit** with layout info |
| `backend` | `Backend` | Yes | - | Backend to show layout on |
| `view` | `str` | No | `'virtual'` | 'virtual' or 'physical' view |

### Returns
`matplotlib.figure.Figure` - Visualization of qubit mapping on device

### âš ï¸ CRITICAL TRAP: Must Use TRANSPILED Circuit

```python
# âŒ WRONG - original circuit has no layout
plot_circuit_layout(qc, backend)  # ERROR!

# âœ… CORRECT - transpiled circuit has layout
transpiled = transpile(qc, backend)
plot_circuit_layout(transpiled, backend)  # Works!
```

### Why This Matters
After transpilation, logical qubits (0, 1, 2...) are mapped to physical qubits on the device. This visualization shows:
- Which physical qubits your algorithm uses
- Why SWAP gates may have been inserted
- How to optimize qubit placement

````python
# ============================================================
# plot_circuit_layout() - Transpilation Mapping
# ============================================================

from qiskit import transpile
from qiskit.visualization import plot_circuit_layout

# Create a circuit
qc_layout = QuantumCircuit(3)
qc_layout.h(0)
qc_layout.cx(0, 1)
qc_layout.cx(1, 2)

# Transpile to backend - this assigns physical qubits
transpiled = transpile(qc_layout, backend)

# Show which physical qubits were assigned
print("Logical to Physical qubit mapping:")
print(transpiled.layout.initial_layout)

# Visualize the layout on the device
plot_circuit_layout(transpiled, backend)
````

````python
# ============================================================
# âš ï¸ TRAP: Original circuit has NO layout info!
# ============================================================

print("âš ï¸ TRAP: Must use TRANSPILED circuit for plot_circuit_layout")
print("="*60)

# âŒ WRONG - Original circuit has no layout
qc_original = QuantumCircuit(3)
qc_original.h(0)
qc_original.cx(0, 1)

print("Original circuit layout:", qc_original.layout)
print("  â†’ None! No transpilation = no layout info")

# âœ… CORRECT - Transpiled circuit HAS layout
from qiskit import transpile
qc_transpiled = transpile(qc_original, backend)

print(f"\nTranspiled circuit layout: {qc_transpiled.layout}")
print("  â†’ Layout object with qubit mapping!")

# Demonstration of the error
print("\nâš ï¸ EXAM TIP: If you see AttributeError or empty visualization,")
print("   check that you're using the TRANSPILED circuit, not original!")
````

## Dynamic Circuit Visualization

````python
# ============================================================
# if_test - Conditional gates
# ============================================================

from qiskit.circuit import QuantumRegister, ClassicalRegister

qr = QuantumRegister(1, 'q')
cr = ClassicalRegister(1, 'c')
qc_dyn = QuantumCircuit(qr, cr)

# Measure and conditionally apply X
qc_dyn.h(qr[0])
qc_dyn.measure(qr[0], cr[0])

with qc_dyn.if_test((cr[0], 1)):  # If measured 1
    qc_dyn.x(qr[0])                # Apply X gate
    
qc_dyn.measure(qr[0], cr[0])

# Best style for dynamic circuits
qc_dyn.draw(output='mpl', style='iqp')
````

````python
# ============================================================
# if-else Control Flow
# ============================================================

qr2 = QuantumRegister(2, 'q')
cr2 = ClassicalRegister(2, 'c')
qc_ifelse = QuantumCircuit(qr2, cr2)

qc_ifelse.h(qr2[0])
qc_ifelse.measure(qr2[0], cr2[0])

# If-else: if c0==1 apply H, else apply X
with qc_ifelse.if_test((cr2[0], 1)) as else_:
    qc_ifelse.h(qr2[1])
with else_:
    qc_ifelse.x(qr2[1])
    
qc_ifelse.measure(qr2[1], cr2[1])
qc_ifelse.draw(output='mpl', style='iqp')
````

## ğŸ“‹ Code Patterns

````python
# ============================================================
# PATTERN: Publication-quality circuit
# ============================================================

def draw_publication_quality(qc, filename=None):
    """Draw circuit with publication-quality settings."""
    return qc.draw(
        output='mpl',
        reverse_bits=True,   # MSB on top (physics convention)
        fold=-1,             # Never fold
        scale=1.3,           # Slightly larger
        idle_wires=False,    # Hide unused
        style='bw',          # Black & white for papers
        filename=filename
    )

# Demo
qc_ghz = QuantumCircuit(4, 4)
qc_ghz.h(0)
for i in range(3):
    qc_ghz.cx(i, i+1)
qc_ghz.barrier()
qc_ghz.measure(range(4), range(4))

draw_publication_quality(qc_ghz)
````

````python
# ============================================================
# PATTERN: Quick debug print
# ============================================================

def debug_circuit(qc, label="Circuit"):
    """Quick circuit debug with state info."""
    print(f"\n{'='*40}")
    print(f"{label}")
    print(f"Qubits: {qc.num_qubits}, Classical bits: {qc.num_clbits}")
    print(f"Depth: {qc.depth()}, Gate count: {qc.size()}")
    print(f"{'='*40}")
    print(qc.draw())

# Demo
debug_circuit(qc_ghz, "GHZ State Circuit")
````

## ğŸ”§ Part 2: Circuit Drawing Options (EXAM TESTED!)

### Option 1: reverse_bits - Qubit Ordering

**EXAM CRITICAL**: This changes which qubit appears on top!

- **Default (`reverse_bits=False`)**: q_0 on top (Qiskit convention)
- **Reversed (`reverse_bits=True`)**: q_n on top (MSB first, classical convention)

````python
# Create 3-qubit circuit for clear demonstration
qc3 = QuantumCircuit(3)
qc3.h(0)
qc3.cx(0, 1)
qc3.cx(1, 2)

print("Default (reverse_bits=False): q_0 on TOP")
print(qc3.draw())
print("\n" + "="*50)

print("\nReversed (reverse_bits=True): q_2 on TOP (MSB first)")
print(qc3.draw(reverse_bits=True))
````

### âš ï¸ EXAM TRAP #3: reverse_bits Significance

**Exam Question Pattern**: "What does reverse_bits=True do?"

**Answer**: Puts MSB (Most Significant Bit) on top, matching classical bit ordering conventions.

**Visual Comparison**:
```
Default (reverse_bits=False):       Reversed (reverse_bits=True):
     q_0: â”€â”€â”€  â† Top (LSB)              q_2: â”€â”€â”€  â† Top (MSB)
     q_1: â”€â”€â”€                           q_1: â”€â”€â”€
     q_2: â”€â”€â”€  â† Bottom (MSB)           q_0: â”€â”€â”€  â† Bottom (LSB)
```

**When to use `reverse_bits=True`**:
- âœ… Matching classical register ordering
- âœ… Big-endian representation
- âœ… When comparing with classical algorithms

**Default is False** (q_0 on top) - Qiskit's natural convention!

## ğŸ“Š Part 4: Understanding `counts` (EXAM CRITICAL!)

Before visualizing measurement results, you need to understand **counts** â€” the output from running quantum circuits.

### ğŸ“– Definition

**`counts`** is a Python dictionary that maps measurement outcome strings to the number of times each outcome was observed during circuit execution.

### Structure

```python
counts = {'00': 512, '11': 488}  # Example for Bell state with 1000 shots
#          â”‚     â”‚
#          â”‚     â””â”€â”€ Number of times this outcome occurred
#          â””â”€â”€ Measurement outcome (binary string)
```

### Key Concepts

| Term | Description |
|------|-------------|
| **Keys** | Binary strings representing measured qubit states (`'00'`, `'01'`, etc.) |
| **Values** | Integers showing how many times each state was measured |
| **Shots** | Total number of circuit executions (sum of all values) |
| **Probabilities** | Divide each count by total shots |

### Why It Matters

- Quantum measurements are **probabilistic** â†’ need multiple shots to see distribution
- `counts` reveals the **probability distribution** of quantum states
- Essential for analyzing algorithm results and verifying circuits

### âš ï¸ EXAM TRAP #4: plot_histogram vs Sampler Result

```python
from qiskit.visualization import plot_histogram

# âŒ WRONG: Can't plot result object directly
plot_histogram(result)  

# âœ… CORRECT: Extract counts first
counts = result[0].data.meas.get_counts()
plot_histogram(counts)
```

## ğŸ§ª Part 5: Circuit Decomposition Visualization

### Viewing Gate Decompositions

**Purpose**: See how complex gates are built from basic gates

````python
# Demonstrate counts structure
from qiskit import QuantumCircuit
from qiskit.primitives import StatevectorSampler

# Create Bell state circuit
qc_bell = QuantumCircuit(2, 2)
qc_bell.h(0)
qc_bell.cx(0, 1)
qc_bell.measure([0, 1], [0, 1])

print("Bell State Circuit:")
print(qc_bell.draw())

# Run and get counts
sampler = StatevectorSampler()
job = sampler.run([qc_bell], shots=1000)
counts = job.result()[0].data.c.get_counts()

print("\nğŸ“Š Counts Dictionary:")
print(f"counts = {counts}")

print("\nğŸ“ˆ Accessing counts:")
total_shots = sum(counts.values())
print(f"Total shots: {total_shots}")
print(f"Most common outcome: {max(counts, key=counts.get)}")
if '00' in counts:
    print(f"Probability of '00': {counts['00'] / total_shots:.2%}")
````

## ğŸ”€ Dynamic Circuit Visualization (Classical Feedforward)

**Dynamic circuits** use mid-circuit measurements to control subsequent operations. The visualization shows conditional operations with special notation.

### Key Concepts:
- **Black filled circle** = conditioning on value 1
- **White open circle** = conditioning on value 0
- **`if_test`** uses context manager syntax (`with`)

````python
# Simple if_test control flow
from qiskit.circuit import QuantumCircuit, QuantumRegister, ClassicalRegister

qubits = QuantumRegister(1, 'q')
clbits = ClassicalRegister(1, 'c')
circuit = QuantumCircuit(qubits, clbits)
(q0,) = qubits
(c0,) = clbits

# Measure and conditionally apply X gate
circuit.h(q0)
circuit.measure(q0, c0)
with circuit.if_test((c0, 1)):  # If c0 == 1, apply X
    circuit.x(q0)
circuit.measure(q0, c0)

# Visualize dynamic circuit - shows conditional blocks
circuit.draw("mpl")
````

### If-Else Control Flow Visualization

````python
# If-else control flow structure
from qiskit.circuit import QuantumCircuit, QuantumRegister, ClassicalRegister

qubits = QuantumRegister(2, 'q')
clbits = ClassicalRegister(2, 'c')
circuit = QuantumCircuit(qubits, clbits)
(q0, q1) = qubits
(c0, c1) = clbits

circuit.h(q0)
circuit.measure(q0, c0)

# If-else structure: if c0 == 1 apply H, else apply X
with circuit.if_test((c0, 1)) as else_:
    circuit.h(q1)      # If c0 == 1: apply H to q1
with else_:
    circuit.x(q1)      # Else: apply X to q1
circuit.measure(q1, c1)

# Shows if-else blocks clearly
circuit.draw("mpl")
````

### Multi-bit Conditional Visualization

````python
# Multi-bit conditional - condition on multiple classical bits
from qiskit.circuit import QuantumCircuit, QuantumRegister, ClassicalRegister

qubits = QuantumRegister(3, 'q')
clbits = ClassicalRegister(3, 'c')
circuit = QuantumCircuit(qubits, clbits)
(q0, q1, q2) = qubits

circuit.h([q0, q1])
circuit.measure(q0, clbits[0])
circuit.measure(q1, clbits[1])

# Condition on multiple classical bits
# 0b001 means c0=1, c1=0, c2=0
with circuit.if_test((clbits, 0b001)):
    circuit.x(q2)
circuit.measure(q2, clbits[2])

# Shows multi-bit conditional notation
circuit.draw("mpl")
````

### Best Style for Dynamic Circuits: IQP

Use `style="iqp"` for optimized dynamic circuit visualization.

````python
# IQP style is optimized for dynamic circuits with control flow
circuit.draw(
    output="mpl",
    style="iqp",           # Best for control flow visualization
    idle_wires=False,      # Hide unused wires for clarity
    fold=-1                # Don't fold (show full circuit)
)
````
