## ğŸ¨ Part 1: Superposition vs Entanglement Visualization

This is one of the **most important comparisons** for the exam! Let's visualize the difference between:
- **Superposition**: Independent qubits each in superposition
- **Entanglement**: Correlated qubits that decide together

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# SUPERPOSITION: 2 independent qubits
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
qc_superposition = QuantumCircuit(2, 2)
qc_superposition.h(0)  # Qubit 0 in superposition
qc_superposition.h(1)  # Qubit 1 in superposition (INDEPENDENTLY)
qc_superposition.measure([0, 1], [0, 1])

print("SUPERPOSITION Circuit (2 independent qubits):")
print(qc_superposition.draw())
print("\nExpected: 4 outcomes equally likely (00, 01, 10, 11)")
print("Each qubit decides INDEPENDENTLY")
````

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ENTANGLEMENT: 2 correlated qubits (Bell state)
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
qc_entanglement = QuantumCircuit(2, 2)
qc_entanglement.h(0)       # Qubit 0 in superposition
qc_entanglement.cx(0, 1)   # Entangle with qubit 1
qc_entanglement.measure([0, 1], [0, 1])

print("ENTANGLEMENT Circuit (2 correlated qubits):")
print(qc_entanglement.draw())
print("\nExpected: Only 2 outcomes (00, 11)")
print("Qubits decide TOGETHER - perfectly correlated!")
````

````python
# Run both circuits and compare histograms
sampler = StatevectorSampler()

# Run superposition circuit
job_super = sampler.run([qc_superposition], shots=1000)
counts_super = job_super.result()[0].data.c.get_counts()

# Run entanglement circuit
job_entangled = sampler.run([qc_entanglement], shots=1000)
counts_entangled = job_entangled.result()[0].data.c.get_counts()

print("Superposition counts:", counts_super)
print("Entanglement counts:", counts_entangled)

# Compare side by side
plot_histogram(
    [counts_super, counts_entangled],
    legend=['Superposition (independent)', 'Entanglement (correlated)'],
    title='Superposition vs Entanglement: See the Difference!',
    figsize=(12, 5)
)
````

### ğŸ” Key Observation

**Superposition** (4 bars):
- `00`: ~25%, `01`: ~25%, `10`: ~25%, `11`: ~25%
- Each qubit flips independently â†’ all 4 combinations possible

**Entanglement** (2 bars):
- `00`: ~50%, `11`: ~50%
- `01`: 0%, `10`: 0%
- Qubits are correlated â†’ only matching pairs appear!

**Exam Pattern**: "Given this histogram, is the state superposed or entangled?"

## ğŸŒ Part 2: All State Visualizations (Statevector)

Let's visualize the Bell state using all 5 state visualization functions:

````python
# Create Bell state circuit (NO measurements for Statevector!)
qc_bell = QuantumCircuit(2)
qc_bell.h(0)
qc_bell.cx(0, 1)

# Get the Statevector
state = Statevector(qc_bell)
print("Bell State: (|00âŸ© + |11âŸ©)/âˆš2")
print(f"Statevector: {state.data}")
print("\nNote: Circuit has NO measurements (required for Statevector)")
````

### Visualization 1: `plot_state_city()` - 3D Amplitude Bars

**a) WHEN TO USE**: When you need precise numerical values of state amplitudes

**b) WHERE IT HELPS**: Debugging algorithms, verifying state preparation

**c) PURPOSE**: Displays complex amplitudes as 3D bars (real & imaginary parts)

````python
plot_state_city(state, title="Bell State - City Plot (Amplitudes)")
````

### Visualization 2: `plot_state_qsphere()` - Probability + Phase Sphere

**a) WHEN TO USE**: When analyzing multi-qubit states where phase relationships matter

**b) WHERE IT HELPS**: Understanding superposition patterns, detecting phase errors

**c) PURPOSE**: Shows probability (marker size) and phase (color) on spherical surface

````python
plot_state_qsphere(state, title="Bell State - Q-Sphere (Probability + Phase)")
````

### Visualization 3: `plot_bloch_multivector()` - Individual Qubit Spheres

**a) WHEN TO USE**: When examining how each qubit behaves independently

**b) WHERE IT HELPS**: Detecting entanglement (mixed states appear at center)

**c) PURPOSE**: Visualizes each qubit's reduced state on its own Bloch sphere

âš ï¸ **Note**: For entangled states, arrows appear at CENTER (mixed state)!

````python
plot_bloch_multivector(state, title="Bell State - Bloch Spheres (Per Qubit)")
print("âš ï¸ Arrows at CENTER = entangled state (qubits appear mixed when viewed alone)")
````

### Visualization 4: `plot_state_hinton()` - Density Matrix

**a) WHEN TO USE**: When analyzing density matrices, entanglement, or mixed states

**b) WHERE IT HELPS**: Identifying off-diagonal coherences, spotting decoherence

**c) PURPOSE**: Displays density matrix elements as squares (size = magnitude, color = sign)

````python
plot_state_hinton(state, title="Bell State - Hinton Diagram (Density Matrix)")
````

### Visualization 5: `plot_state_paulivec()` - Pauli Decomposition

**a) WHEN TO USE**: When working with Hamiltonians or Pauli-based decompositions

**b) WHERE IT HELPS**: VQE algorithms, quantum chemistry, error analysis

**c) PURPOSE**: Shows expectation values of all Pauli operator combinations (II, IX, IY, IZ, XI, XX, ...)

````python
plot_state_paulivec(state, title="Bell State - Pauli Vector Decomposition")
````

## ğŸ“Š Part 3: Measurement Visualization (StatevectorSampler)

Now let's visualize the **same Bell state** using measurement-based visualizations:

````python
# Create Bell state WITH measurements (required for StatevectorSampler!)
qc_bell_meas = QuantumCircuit(2, 2)
qc_bell_meas.h(0)
qc_bell_meas.cx(0, 1)
qc_bell_meas.measure([0, 1], [0, 1])  # MUST have measurements!

# Run the sampler
sampler = StatevectorSampler()
job = sampler.run([qc_bell_meas], shots=1000)
counts = job.result()[0].data.c.get_counts()

print("Circuit with measurements:")
print(qc_bell_meas.draw())
print(f"\nCounts: {counts}")
````

### `plot_histogram()` - Measurement Counts

Shows the **raw counts** (integer frequencies) from measurement:

````python
plot_histogram(counts, title="Bell State - Histogram (Measurement Counts)")
````

### `plot_distribution()` - Normalized Probabilities

Shows **normalized probabilities** (values sum to 1):

````python
plot_distribution(counts, title="Bell State - Distribution (Normalized Probabilities)")
````

## âš ï¸ Part 4: Common Errors - Wrong Tool + Wrong Visualization

Let's demonstrate the errors you'll see if you mix up tools and visualizations:

````python
# âŒ ERROR 1: Using state visualization with counts
print("âŒ ERROR: Using plot_state_city with counts dictionary")
print("   plot_state_city({'00': 500, '11': 500})")
print("   â†’ TypeError: Expects Statevector, not dict!")

print("\nâœ… CORRECT: Use plot_histogram for counts")
print("   plot_histogram({'00': 500, '11': 500})")
````

````python
# âŒ ERROR 2: Using histogram with Statevector
print("âŒ ERROR: Using plot_histogram with Statevector")
print("   plot_histogram(Statevector(qc))")
print("   â†’ TypeError: Expects dict, not Statevector!")

print("\nâœ… CORRECT: Use state visualizations for Statevector")
print("   plot_state_city(Statevector(qc))")

# âŒ ERROR 3: Statevector on circuit WITH measurements
print("\nâŒ WARNING: Statevector on circuit WITH measurements")
print("   Works but gives post-measurement state (not what you want!)")

# âŒ ERROR 4: StatevectorSampler WITHOUT measurements
print("\nâŒ ERROR: StatevectorSampler WITHOUT measurements")
print("   No counts to retrieve! Always add measure() first.")
````

## ğŸ¯ Part 5: Complete Workflow - Both Tools, All Visualizations

Let's see a complete analysis of a GHZ state using BOTH Statevector and StatevectorSampler:

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# COMPLETE GHZ STATE ANALYSIS
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•

# Step 1: Create GHZ circuit (NO measurements for Statevector)
qc_ghz = QuantumCircuit(3)
qc_ghz.h(0)
qc_ghz.cx(0, 1)
qc_ghz.cx(1, 2)

print("="*60)
print("GHZ STATE: (|000âŸ© + |111âŸ©)/âˆš2")
print("="*60)

print("\nğŸ“‹ 1. CIRCUIT DIAGRAM:")
print(qc_ghz.draw())

# Step 2: Get Statevector (exact quantum state)
state_ghz = Statevector(qc_ghz)
print("\nğŸ”¬ 2. STATEVECTOR (exact amplitudes):")
print(state_ghz.data)
````

````python
# Step 3: All STATE visualizations (using Statevector)
print("ğŸŒ 3. STATE VISUALIZATIONS (Statevector - no measurements):")

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

plt.subplot(2, 2, 1)
plot_state_city(state_ghz, ax=plt.gca(), title="City Plot")

plt.subplot(2, 2, 2)
plot_state_qsphere(state_ghz, ax=plt.gca())
plt.title("Q-Sphere")

# Note: Bloch and Hinton need separate plots
plt.tight_layout()
plt.show()

print("\nğŸ”® Bloch Spheres (each qubit separately):")
plot_bloch_multivector(state_ghz)
plt.show()
````

````python
# Step 4: Add measurements and run StatevectorSampler
qc_ghz_meas = qc_ghz.copy()
qc_ghz_meas.measure_all()

sampler = StatevectorSampler()
job = sampler.run([qc_ghz_meas], shots=1000)
counts_ghz = job.result()[0].data.meas.get_counts()

print("ğŸ“Š 4. MEASUREMENT VISUALIZATION (StatevectorSampler - needs measurements):")
print(f"   Counts: {counts_ghz}")
print("   Expected: ~50% for 000, ~50% for 111 (perfect 3-qubit correlation!)")
````

````python
plot_histogram(counts_ghz, title="GHZ State - Measurement Results", bar_labels=True)
````
