### Analogy: The Musical Score

A quantum circuit diagram is remarkably similar to a **musical score**:
- **Time**: Flows from left to right
- **Staff Lines**: Each line represents a qubit (like an instrument)

- **Barriers**: Like measure lines, organizing the music into sections

## Setup

Import all required libraries:

````python
# === SETUP ===
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.primitives import StatevectorSampler
from qiskit.visualization import (
    plot_histogram, plot_distribution,
    plot_bloch_multivector, plot_state_city,
    plot_state_qsphere, plot_state_hinton,
    plot_state_paulivec
)
import matplotlib.pyplot as plt

sampler = StatevectorSampler()
print("Setup complete")
````

---

## 1. Superposition vs Entanglement (EXAM CRITICAL)

**Pattern recognition**: Given a histogram, identify the state type

````python
# SUPERPOSITION: Independent qubits
qc_super = QuantumCircuit(2, 2)
qc_super.h(0)
qc_super.h(1)  # Each qubit INDEPENDENTLY in superposition
qc_super.measure([0, 1], [0, 1])

# ENTANGLEMENT: Correlated qubits
qc_entang = QuantumCircuit(2, 2)
qc_entang.h(0)
qc_entang.cx(0, 1)  # Qubits decide TOGETHER
qc_entang.measure([0, 1], [0, 1])

print("Superposition:", qc_super.draw(output='text'))
print("\nEntanglement:", qc_entang.draw(output='text'))
````

````python
# Run both and compare
job = sampler.run([qc_super, qc_entang], shots=1000)
counts_super = job.result()[0].data.c.get_counts()
counts_entang = job.result()[1].data.c.get_counts()

print("Superposition counts:", counts_super)
print("Entanglement counts:", counts_entang)

plot_histogram(
    [counts_super, counts_entang],
    legend=['Superposition', 'Entanglement'],
    title='Pattern: 4 bars = Superposition | 2 bars = Entanglement',
    figsize=(10, 4)
)
````

**EXAM KEY**:
- **4 equal bars** (00, 01, 10, 11) = Superposition (independent qubits)
- **2 correlated bars** (00, 11 only) = Entanglement (Bell state)

````python
# 4. plot_state_hinton - Density matrix squares
plot_state_hinton(state, title='Hinton: Density matrix (size=magnitude)')
````

## Verification: Tool Compatibility

````python
# Verify compatibility rules
def test_compatibility():
    # Circuit without measurements
    qc = QuantumCircuit(2)
    qc.h(0)
    qc.cx(0, 1)
    
    # Statevector works
    state = Statevector(qc)
    assert state is not None
    
    # All state visualizations work with Statevector
    assert plot_state_city(state) is not None
    assert plot_state_qsphere(state) is not None
    assert plot_bloch_multivector(state) is not None
    assert plot_state_hinton(state) is not None
    assert plot_state_paulivec(state) is not None
    plt.close('all')
    
    # Circuit with measurements
    qc.measure_all()
    job = sampler.run([qc], shots=100)
    counts = job.result()[0].data.meas.get_counts()
    
    # Histogram works with counts
    assert plot_histogram(counts) is not None
    plt.close('all')
    
    return True

print("All compatibility tests passed:", test_compatibility())
````

## ğŸ¨ Part 1: Superposition vs Entanglement Visualization

This is one of the **most important comparisons** for the exam! Let's visualize the difference between:
- **Superposition**: Independent qubits each in superposition
- **Entanglement**: Correlated qubits that decide together

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# SUPERPOSITION: 2 independent qubits
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
qc_superposition = QuantumCircuit(2, 2)
qc_superposition.h(0)  # Qubit 0 in superposition
qc_superposition.h(1)  # Qubit 1 in superposition (INDEPENDENTLY)
qc_superposition.measure([0, 1], [0, 1])

print("SUPERPOSITION Circuit (2 independent qubits):")
print(qc_superposition.draw())
print("\nExpected: 4 outcomes equally likely (00, 01, 10, 11)")
print("Each qubit decides INDEPENDENTLY")
````

````python
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
# ENTANGLEMENT: 2 correlated qubits (Bell state)
# â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•
qc_entanglement = QuantumCircuit(2, 2)
qc_entanglement.h(0)       # Qubit 0 in superposition
qc_entanglement.cx(0, 1)   # Entangle with qubit 1
qc_entanglement.measure([0, 1], [0, 1])

print("ENTANGLEMENT Circuit (2 correlated qubits):")
print(qc_entanglement.draw())
print("\nExpected: Only 2 outcomes (00, 11)")
print("Qubits decide TOGETHER - perfectly correlated!")
````

````python
# Run both circuits and compare histograms
sampler = StatevectorSampler()

# Run superposition circuit
job_super = sampler.run([qc_superposition], shots=1000)
counts_super = job_super.result()[0].data.c.get_counts()

# Run entanglement circuit
job_entangled = sampler.run([qc_entanglement], shots=1000)
counts_entangled = job_entangled.result()[0].data.c.get_counts()

print("Superposition counts:", counts_super)
print("Entanglement counts:", counts_entangled)

# Compare side by side
plot_histogram(
    [counts_super, counts_entangled],
    legend=['Superposition (independent)', 'Entanglement (correlated)'],
    title='Superposition vs Entanglement: See the Difference!',
    figsize=(12, 5)
)
````

### ğŸ” Key Observation

**Superposition** (4 bars):
- `00`: ~25%, `01`: ~25%, `10`: ~25%, `11`: ~25%
- Each qubit flips independently â†’ all 4 combinations possible

**Entanglement** (2 bars):
- `00`: ~50%, `11`: ~50%
- `01`: 0%, `10`: 0%
- Qubits are correlated â†’ only matching pairs appear!

**Exam Pattern**: "Given this histogram, is the state superposed or entangled?"

````python
# Create Bell state circuit (NO measurements for Statevector!)
qc_bell = QuantumCircuit(2)
qc_bell.h(0)
qc_bell.cx(0, 1)

# Get the Statevector
state = Statevector(qc_bell)
print("Bell State: (|00âŸ© + |11âŸ©)/âˆš2")
print(f"Statevector: {state.data}")
print("\nNote: Circuit has NO measurements (required for Statevector)")
````

### Visualization 1: `plot_state_city()` - 3D Amplitude Bars

**a) WHEN TO USE**: When you need precise numerical values of state amplitudes

**b) WHERE IT HELPS**: Debugging algorithms, verifying state preparation

### Visualization 2: `plot_state_qsphere()` - Probability + Phase Sphere

**a) WHEN TO USE**: When analyzing multi-qubit states where phase relationships matter

**b) WHERE IT HELPS**: Understanding superposition patterns, detecting phase errors

**c) PURPOSE**: Shows probability (marker size) and phase (color) on spherical surface

````python
plot_state_qsphere(state, title="Bell State - Q-Sphere (Probability + Phase)")
````

### Visualization 3: `plot_bloch_multivector()` - Individual Qubit Spheres

**a) WHEN TO USE**: When examining how each qubit behaves independently

**b) WHERE IT HELPS**: Detecting entanglement (mixed states appear at center)

**c) PURPOSE**: Visualizes each qubit's reduced state on its own Bloch sphere

âš ï¸ **Note**: For entangled states, arrows appear at CENTER (mixed state)!

**a) WHEN TO USE**: When analyzing density matrices, entanglement, or mixed states

**b) WHERE IT HELPS**: Identifying off-diagonal coherences, spotting decoherence

````python
plot_state_hinton(state, title="Bell State - Hinton Diagram (Density Matrix)")
````

### Visualization 5: `plot_state_paulivec()` - Pauli Decomposition

**a) WHEN TO USE**: When working with Hamiltonians or Pauli-based decompositions

**b) WHERE IT HELPS**: VQE algorithms, quantum chemistry, error analysis

````python
plot_state_paulivec(state, title="Bell State - Pauli Vector Decomposition")
````

---

## âš ï¸ MASTER TRAP DEMONSTRATION: All Critical Errors

These are the MOST COMMON visualization errors on the exam. See each error in action!

````python
# ============================================================
# TRAP 1: qc.draw() default is TEXT, not MPL
# ============================================================
print("âš ï¸ TRAP 1: Default qc.draw() is TEXT output")
print("="*55)

qc_trap = QuantumCircuit(2)
qc_trap.h(0)
qc_trap.cx(0, 1)

# Default is text - NOT a matplotlib figure!
result = qc_trap.draw()
print(f"type(qc.draw()): {type(result)}")
print("  â†’ Returns text string, not figure!")

print("\nâœ… CORRECT: qc.draw('mpl') for graphical output")
print(f"type(qc.draw('mpl')): {type(qc_trap.draw('mpl'))}")
print("  â†’ Returns matplotlib Figure!")
````

````python
# ============================================================
# TRAP 2: Style only works with MPL output
# ============================================================
print("âš ï¸ TRAP 2: style= parameter ignored without output='mpl'")
print("="*55)

# âŒ WRONG - style ignored with text output!
text_styled = qc_trap.draw(style='iqp')
print("qc.draw(style='iqp'):")
print(text_styled)
print("\n  â†’ Style parameter IGNORED! Still text output!")

# âœ… CORRECT - must specify mpl for style to work
print("\nâœ… CORRECT: qc.draw(output='mpl', style='iqp')")
print("  â†’ Style applied to matplotlib figure!")
````

````python
# ============================================================
# TRAP 3: Statevector vs StatevectorSampler - OPPOSITE requirements!
# ============================================================
print("âš ï¸ TRAP 3: Statevector vs StatevectorSampler measurement rules")
print("="*55)

# Setup for demonstration
from qiskit.quantum_info import Statevector
from qiskit.primitives import StatevectorSampler

# Statevector: NO measurements allowed
qc_no_meas = QuantumCircuit(2)
qc_no_meas.h(0)
qc_no_meas.cx(0, 1)

# StatevectorSampler: MUST have measurements
qc_with_meas = QuantumCircuit(2, 2)
qc_with_meas.h(0)
qc_with_meas.cx(0, 1)
qc_with_meas.measure([0, 1], [0, 1])

print("Statevector: Circuit must NOT have measurements")
print(f"  Statevector(qc_no_meas) â†’ Works! âœ…")

print("\nStatevectorSampler: Circuit MUST have measurements")
print(f"  sampler.run([qc_with_meas]) â†’ Works! âœ…")

print("\nâŒ COMMON MISTAKES:")
print("  â€¢ Statevector(circuit_with_measurements) â†’ collapsed state!")
print("  â€¢ sampler.run([circuit_without_measurements]) â†’ no counts!")
````
