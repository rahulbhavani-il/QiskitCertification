## 1. Foundational Circuits

### 1.1 Bell States (Œ¶‚Å∫, Œ¶‚Åª, Œ®‚Å∫, Œ®‚Åª)

#### **a) Purpose**
Bell states are the simplest examples of maximally entangled two-qubit states. They demonstrate quantum entanglement, where measuring one qubit instantly determines the state of the other, regardless of distance.

| State | Circuit | Result |
|-------|---------|--------|
| Œ¶‚Å∫ | H(0), CX(0,1) | (|00‚ü© + |11‚ü©)/‚àö2 |
| Œ¶‚Åª | X(0), H(0), CX(0,1) | (|00‚ü© - |11‚ü©)/‚àö2 |
| Œ®‚Å∫ | H(0), CX(0,1), X(1) | (|01‚ü© + |10‚ü©)/‚àö2 |
| Œ®‚Åª | H(0), CX(0,1), X(1), Z(1) | (|01‚ü© - |10‚ü©)/‚àö2 |

**Bell State Œ¶‚Å∫ (most common):**

````python
qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
qc.draw("mpl")
````

**Bell State Œ¶‚Åª:**

````python
qc = QuantumCircuit(2)
qc.x(0)
qc.h(0)
qc.cx(0, 1)
qc.draw("mpl")
````

**Bell State Œ®‚Å∫:**

````python
qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
qc.x(1)
qc.draw("mpl")
````

**Bell State Œ®‚Åª:**

````python
qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
qc.x(1)
qc.z(1)
qc.draw("mpl")
````

#### **b) Why Important**
- **Foundation of quantum entanglement** - The Bell states are the "Hello World" of quantum computing
- **Building block for most quantum algorithms** - Teleportation, superdense coding, error correction
- **Exam staple** - Appears in 60%+ of Qiskit certification questions
- **Testing quantum hardware** - Used to verify genuine entanglement on quantum devices

#### **c) Additional Information**
- **Measurement outcomes**: Bell state Œ¶‚Å∫ always produces 50% |00‚ü© and 50% |11‚ü© - NEVER |01‚ü© or |10‚ü©
- **Correlations**: ‚ü®ZZ‚ü© = +1, ‚ü®XX‚ü© = +1, ‚ü®YY‚ü© = -1 for Œ¶‚Å∫
- **Exam trap**: Remember the order - H goes on control qubit BEFORE the CNOT
- **Memory aid**: "H-CX makes Bell-X" (Hadamard then CNOT makes entanglement)

### 1.2 GHZ State

#### **a) Purpose**
The GHZ (Greenberger-Horne-Zeilinger) state extends entanglement to 3+ qubits, creating a superposition where all qubits are either all |0‚ü© or all |1‚ü©.

````python
qc = QuantumCircuit(3)
qc.h(0)
qc.cx(0, 1)
qc.cx(0, 2)
qc.draw("mpl")
````

#### **b) Why Important**
- **Multi-qubit entanglement** - Demonstrates scalable entanglement beyond 2 qubits
- **Quantum error correction** - GHZ states are used in error syndrome detection
- **Quantum sensing** - Achieves Heisenberg-limited precision in measurements
- **Tests locality** - Used in Bell inequality experiments with 3+ parties

#### **c) Additional Information**
- **Pattern**: H on first qubit, then cascade of CNOTs from first qubit to all others
- **n-qubit GHZ**: (|00...0‚ü© + |11...1‚ü©)/‚àö2
- **Fragile**: Loss of any qubit destroys entanglement (unlike W state)
- **Measurement**: Always gives all-zeros or all-ones (50/50 probability)

### 1.3 W State

#### **a) Purpose**
The W state is an alternative 3-qubit entangled state to GHZ, represented as (|100‚ü© + |010‚ü© + |001‚ü©)/‚àö3. Unlike GHZ, W states maintain partial entanglement even when one qubit is lost.

````python
import numpy as np

qc = QuantumCircuit(3)
qc.ry(2 * np.arccos(1 / np.sqrt(3)), 0)
qc.cx(0, 1)
qc.ry(2 * np.arccos(1 / np.sqrt(2)), 1)
qc.cx(1, 2)
qc.cx(0, 2)
qc.draw("mpl")
````

#### **b) Why Important**
- **Robust entanglement** - Loss of one qubit leaves remaining pair still entangled
- **Quantum networks** - Useful for quantum communication protocols
- **Contrast to GHZ** - GHZ becomes separable if any qubit is lost, W state doesn't

#### **c) Additional Information**
- **vs GHZ**: GHZ = (|000‚ü© + |111‚ü©)/‚àö2, W = (|100‚ü© + |010‚ü© + |001‚ü©)/‚àö3
- **Measurement**: Always gives exactly one |1‚ü© and rest |0‚ü©s
- **Scaling**: n-qubit W state has equal amplitude for all single-excitation states
- **Property**: Under local operations, W and GHZ cannot be converted to each other
- **Exam tip**: W state is more "distributed" than GHZ state

### 1.4 Superposition Circuit

#### **a) Purpose**
Creates an equal superposition of all computational basis states using Hadamard gates on each qubit. This is the quantum analog of exploring all possibilities simultaneously.

````python
qc = QuantumCircuit(3)
qc.h(0)
qc.h(1)
qc.h(2)
qc.draw("mpl")
````

#### **b) Why Important**
- **Quantum parallelism** - Process 2^n states simultaneously
- **Algorithm initialization** - First step in Grover's, Deutsch-Jozsa, etc.
- **Demonstrates quantum advantage** - Classical computers can't represent this efficiently

#### **c) Additional Information**
- **Measurement**: Each of 2^n outcomes has probability 1/2^n
- **State**: (|0‚ü© + |1‚ü©)‚äón / ‚àö(2^n) = Œ£|x‚ü© / ‚àö(2^n) for all binary x
- **Exam tip**: 3 qubits = 8 equal outcomes, 4 qubits = 16 equal outcomes
- **No entanglement**: Superposition ‚â† entanglement (parallel H gates don't entangle)

### 1.5 Phase Kickback Circuit

#### **a) Purpose**
Demonstrates phase kickback - when a controlled gate is applied to an eigenstate of the target operation, the phase is "kicked back" to the control qubit. This is the secret behind many quantum algorithms.

````python
qc = QuantumCircuit(2)
qc.h(0)
qc.x(1)
qc.cx(0, 1)
qc.h(0)
qc.draw("mpl")
````

#### **b) Why Important**
- **Core mechanism of quantum algorithms** - Grover's, Shor's, phase estimation all use this
- **Explains quantum speedup** - Allows measuring properties of functions
- **Counter-intuitive** - Phase appears on control, not target

#### **c) Additional Information**
- **Mathematical basis**: If U|œà‚ü© = e^(iœÜ)|œà‚ü©, then CU|+‚ü©|œà‚ü© = e^(iœÜ)|+‚ü©|œà‚ü©
- **Key insight**: The "action" happens on the control qubit
- **Memory aid**: "Phase flows upward" - from target to control
- **Related**: Basis of quantum phase estimation algorithm

## 5. Quantum Algorithms

### 5.1 Deutsch-Jozsa Algorithm

The Deutsch-Jozsa algorithm determines whether a function f: {0,1}‚Åø ‚Üí {0,1} is constant or balanced using a single function evaluation.

**Why important:** This algorithm demonstrates quantum advantage over classical algorithms for the first time, showing exponential speedup for certain problems.

### 5.1.1 Deutsch-Jozsa Circuit

````python
# Deutsch-Jozsa Algorithm Circuit (2 qubits)
qc = QuantumCircuit(3, 2)  # 2 input qubits + 1 output qubit, 2 classical bits

# Initialize output qubit to |1‚ü©
qc.x(2)

# Apply Hadamard gates to all qubits
qc.h(range(3))

# Oracle for balanced function (example: f(x) = x‚ÇÄ ‚äï x‚ÇÅ)
qc.cx(0, 2)
qc.cx(1, 2)

# Apply Hadamard gates to input qubits
qc.h(range(2))

# Measure input qubits
qc.measure(range(2), range(2))

qc.draw("mpl")
````

### 5.2 Grover's Algorithm

Grover's algorithm provides quadratic speedup for searching an unsorted database of N items.

**Why important:** This algorithm demonstrates quantum advantage for search problems and forms the basis for many quantum algorithms.

### 5.2.1 Grover's Algorithm Circuit (2 qubits)

````python
# Grover's Algorithm Circuit (2 qubits, searching for |11‚ü©)
qc = QuantumCircuit(3, 2)  # 2 search qubits + 1 ancilla qubit

# Initialize superposition
qc.h(range(2))

# Initialize ancilla qubit to |1‚ü©
qc.x(2)
qc.h(2)

# Oracle for marked state |11‚ü©
qc.ccx(0, 1, 2)

# Diffusion operator
qc.h(range(2))
qc.x(range(2))
qc.ccx(0, 1, 2)
qc.x(range(2))
qc.h(range(2))

# Measure search qubits
qc.measure(range(2), range(2))

qc.draw("mpl")
````

### 5.3 Shor's Algorithm

Shor's algorithm factors large numbers exponentially faster than classical algorithms.

**Why important:** This algorithm threatens classical cryptography and demonstrates the power of quantum computing for number theory problems.

### 5.3.1 Shor's Algorithm - Period Finding Circuit

````python
# Shor's Algorithm - Simplified Period Finding Circuit
# This is a highly simplified version showing the quantum part of Shor's algorithm
# In practice, Shor's algorithm requires many more qubits and complex arithmetic circuits

qc = QuantumCircuit(8, 4)  # 4 counting qubits + 4 target qubits, 4 classical bits

# Initialize counting qubits in superposition
qc.h(range(4))

# Initialize target register to |1‚ü©
qc.x(7)

# Simplified modular exponentiation: a^x mod N
# For demonstration: a=2, N=15, finding period of f(x) = 2^x mod 15
# This is a toy example - real implementation requires quantum arithmetic

# Measure counting register
qc.measure(range(4), range(4))

qc.draw("mpl")
````

### 5.4 Variational Quantum Eigensolver (VQE)

VQE is a hybrid quantum-classical algorithm for finding the ground state energy of molecular systems.

**Why important:** VQE is one of the most promising near-term quantum algorithms for practical applications in quantum chemistry.

### 5.4.1 VQE Ansatz Circuit

````python
# VQE Ansatz Circuit (Hardware-Efficient Ansatz for 2 qubits)
def vqe_ansatz(params):
    qc = QuantumCircuit(2)

    # Layer 1: Single qubit rotations
    qc.ry(params[0], 0)
    qc.ry(params[1], 1)

    # Layer 2: Entangling gates
    qc.cx(0, 1)

    # Layer 3: Single qubit rotations
    qc.ry(params[2], 0)
    qc.ry(params[3], 1)

    return qc

# Example with dummy parameters
params = [np.pi/4, np.pi/3, np.pi/6, np.pi/2]
qc = vqe_ansatz(params)
qc.draw("mpl")
````

### 5.5 Quantum Approximate Optimization Algorithm (QAOA)

QAOA is designed to solve combinatorial optimization problems on near-term quantum devices.

**Why important:** QAOA provides a general framework for solving optimization problems and is one of the most studied variational quantum algorithms.

### 5.5.1 QAOA Circuit (Max-Cut Problem)

````python
# QAOA Circuit for Max-Cut on a triangle graph
def qaoa_circuit(gamma, beta):
    qc = QuantumCircuit(3)

    # Initial superposition
    qc.h(range(3))

    # Problem Hamiltonian (cost operator) - Max-Cut on triangle
    qc.rz(gamma, 0)
    qc.rz(gamma, 1)
    qc.rz(gamma, 2)

    # Entangling gates for the triangle edges
    qc.cx(0, 1)
    qc.cx(1, 2)
    qc.cx(2, 0)

    # Mixer Hamiltonian
    qc.rx(2*beta, 0)
    qc.rx(2*beta, 1)
    qc.rx(2*beta, 2)

    return qc

# Example with dummy parameters
gamma, beta = np.pi/4, np.pi/6
qc = qaoa_circuit(gamma, beta)
qc.draw("mpl")
````

## 8. Dynamic Circuits

### 8.1 Mid-Circuit Measurement with if_test()

#### **a) Purpose**
Measure qubit mid-circuit and apply conditional operations based on result.

**Why important:** Error correction, adaptive algorithms, modern syntax (replaces deprecated c_if()).

#### **c) Additional Information**
- **Syntax**: `with qc.if_test((classical_bit, expected_value)):`
- **DEPRECATED**: Do NOT use `c_if()` - it's deprecated!
- **Hardware**: Requires dynamic circuits support

````python
# Mid-Circuit Measurement with if_test()
qc = QuantumCircuit(2, 1)
qc.h(0)
qc.measure(0, 0)

# Conditional operation based on measurement result
with qc.if_test((0, 1)):  # if classical bit 0 == 1
    qc.x(1)

qc.draw("mpl")
````

### 8.2 for_loop Circuit

#### **a) Purpose**
Runtime loop that executes a fixed number of iterations on quantum hardware.

**Why important:** Efficiency, single instruction for repeated operations, part of real-time control flow.

#### **c) Additional Information**
- **vs Python loop**: Python loop unrolls at compile time
- **for_loop**: Preserved as loop structure for hardware
- **Use case**: Repeated measurements, iterative algorithms

````python
# for_loop Circuit (runtime loop on hardware)
from qiskit.circuit import QuantumCircuit

qc = QuantumCircuit(1)

# Runtime loop: H applied 3 times on hardware
with qc.for_loop(range(3)):
    qc.h(0)

qc.draw("mpl")
````

### 8.3 Reset Operations

#### **a) Purpose**
Mid-circuit reset operations that return a qubit to |0‚ü© state regardless of its current state. Essential for qubit reuse and error correction.

**Why important:** Qubit reuse, error correction, dynamic circuits, resource optimization.

#### **c) Additional Information**
- **Implementation**: Hardware may use measurement + conditional X gate
- **Timing**: Reset operations have finite duration (not instantaneous)
- **vs measurement**: Measurement reads state, reset changes state to |0‚ü©
- **Syntax**: `qc.reset(qubit_index)`
- **After reset**: Qubit is in |0‚ü© state, can be used for new computation

````python
# Reset Operations (mid-circuit reset)
qc = QuantumCircuit(1)
qc.h(0)  # Put qubit in superposition
qc.reset(0)  # Reset to |0‚ü© regardless of current state
qc.h(0)  # Can now use the qubit again

qc.draw("mpl")
````

### 9.1 Optimization Level Comparison

#### **a) Purpose**
Compare transpilation results at different optimization levels (0-3).

**Why important:** Performance tuning, higher levels = more optimization, trade-off with compilation time.

#### **c) Additional Information**
- **Level 0**: No optimization, just mapping
- **Level 1**: Light optimization

## 11. Error Correction Circuits

### 11.1 3-Qubit Bit Flip Code

#### **a) Purpose**
Simplest quantum error correction code that protects against single bit-flip (X) errors by encoding one logical qubit into three physical qubits.

**Why important:** Quantum error correction, redundancy, syndrome measurement.

#### **c) Additional Information**
- **Encoding**: |0‚ü© ‚Üí |000‚ü©, |1‚ü© ‚Üí |111‚ü©
- **Protects against**: Single bit-flip error (X error on any one qubit)
- **Syndrome table**:
  - `00`: No error
  - `01`: Error on qubit 3
  - `10`: Error on qubit 1
  - `11`: Error on qubit 2

````python
# 3-Qubit Bit Flip Code
# Encoding: |0‚ü© ‚Üí |000‚ü©, |1‚ü© ‚Üí |111‚ü©
qc = QuantumCircuit(5, 2)  # 3 data + 2 syndrome qubits, 2 classical bits

# Encode logical qubit (assume starts as |0‚ü©)
qc.cx(0, 1)
qc.cx(0, 2)

# Syndrome measurement (parity checks)
qc.cx(0, 3)  # syndrome 1: parity of qubits 0,1
qc.cx(1, 3)
qc.cx(1, 4)  # syndrome 2: parity of qubits 1,2
qc.cx(2, 4)

# Measure syndromes
qc.measure(3, 0)
qc.measure(4, 1)

qc.draw("mpl")
````

### 11.2 Syndrome Measurement

#### **a) Purpose**
Measure error syndromes without collapsing the logical quantum state. Key technique for quantum error correction.

**Why important:** Non-destructive error detection, real-time feedback, stabilizer codes foundation.

#### **c) Additional Information**
- **Parity check**: Syndrome measures whether even or odd number of errors
- **Ancilla reset**: Syndrome qubits typically reset between rounds
- **Stabilizer codes**: Syndrome measurement is the foundation

````python
# Syndrome Measurement (parity check without destroying data)
qc = QuantumCircuit(4, 1)  # 2 data qubits + 1 syndrome qubit + 1 classical bit

# Data qubits (assume they may have errors)
# Syndrome measurement: parity of data qubits 0 and 1
qc.cx(0, 2)  # CNOT from data to syndrome
qc.cx(1, 2)  # If even parity: syndrome = |0‚ü©, if odd parity: syndrome = |1‚ü©

# Measure syndrome (non-destructive for data qubits)
qc.measure(2, 0)

qc.draw("mpl")
````
